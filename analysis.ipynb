```python
def trend_following(ser: pd.Series, momentum: float) -> float:
    trend = 0.0
    indice = 0
    owned = False
    value = 1.0

    np_array = ser.to_numpy()
    iterator = enumerate(np_array)
    next(iterator)

    for i, elem in iterator:
        if trend > 0.0:
            owned = True
            value += elem
        else:
            owned = False

        trend = momentum * float(elem) + (1 - momentum) * trend

    return value
```

```python
close_diffs_wig20 = [df.diff()["Close"] for df in dfs_wig20]
momentum_vals = np.linspace(0.01, 1.0, num=100)

returns = []

for momentum in momentum_vals:
    total = 0.0
    for i, df in enumerate(close_diffs_wig20):
        tf = trend_following(df, momentum)
        total += tf

    returns.append(total)

plt.imshow(np.expand_dims(returns, 0))

max(returns)
```

```python
def std_decision(ser: pd.Series, mean: float, std: float, dev: float) -> float:
    value = 1.0
    owned = False
    
    np_array = ser.to_numpy()
    iterator = enumerate(np_array)
    next(iterator)

    for i, elem in iterator:
        if owned:
            value += elem

        if elem >= mean + dev*std:
            owned = True
        elif elem <= mean - dev*std:
            owned = False

    return value
```

```python
def correlation_buying(ticker: str, tickers: list, values: pd.DataFrame, corr: pd.DataFrame) -> float:
    corrs = corr[ticker]
    data = values

    iterators = {}
    for t in tickers:
        iterator = enumerate(data[tickers.index(t)])
        next(iterator)
        
        iterators[t] = iterator

    value = 1.0
    owned = False
    for i, price in iterators[ticker]:
        price_norm = (price - mean[ticker]) / std[ticker]
        synthetic_price = sum([corrs[t]*(next(iterators[t])[1] - mean[t]) / std[t] for t in tickers if t!=ticker]) / sum(corrs)

        if owned:
            value += price

        if synthetic_price > price_norm:
            owned = True
        else:
            owned = False

    return value
```
